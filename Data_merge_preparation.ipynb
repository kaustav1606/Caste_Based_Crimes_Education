```python
import pandas as pd
# Load the data
data = pd.read_excel('All_States_AssemblyElect.xlsx')  # Adjust this to your specific dataset path
```

```python
data
```

```python
data_2009 = data[data['Year']== 2008]
data_2009 = data_2009[(data_2009['State_Name']== 'Madhya_Pradesh')]
data_2009
```

```python
winners_2009 = data_2009[data_2009['Position'] == 1]
runners_up_2009 = data_2009[data_2009['Position'] == 2]
# Step 3: Merge Winners and Runners-Up
merged_results_2009 = winners_2009.merge(
    runners_up_2009,
    on=['District_Name', 'Constituency_No', 'State_Name'],
    suffixes=('_winner', '_runnerup')
)

merged_results_2009['adjusted_margin'] = merged_results_2009['Margin_Percentage_winner']
merged_results_2009.loc[
    merged_results_2009['educdum_winner'] == 0, 'adjusted_margin'
] *= -1
```

```python
# Step 4: Identify Grad-NonGrad Combinations in 2009
grad_non_grad_combination_2009 = (
    (merged_results_2009['educdum_winner'] == 1) & (merged_results_2009['educdum_runnerup'] == 0)
) | (
    (merged_results_2009['educdum_winner'] == 0) & (merged_results_2009['educdum_runnerup'] == 1)
)
```

```python
# Step 5: Select Constituencies with Grad-NonGrad Combinations in 2009
grad_non_grad_constituencies_2009 = merged_results_2009[grad_non_grad_combination_2009]
```

```python
pd.set_option('display.max_columns', 100)
grad_non_grad_constituencies_2009.to_excel('grad_non_grad_2008_new.xlsx', index=False)
```

```python
grad_non_grad_constituencies_2009['abs_adjusted_margin'] = abs(grad_non_grad_constituencies_2009['adjusted_margin'])
# Group by district and sort by absolute value of the adjusted margin
sorted_results = grad_non_grad_constituencies_2009.groupby(['District_Name', 'State_Name']).apply(
    lambda group: group.sort_values(by='abs_adjusted_margin', ascending=True)
)

# Reset multi-index to get a clean DataFrame
sorted_results.reset_index(drop=True, inplace=True)
```

### margin variable

```python
df_1 = pd.read_excel('grad_non_grad_2008_new.xlsx')
#df_2 = pd.read_excel('grad_non_grad_2009.xlsx',sheet_name='reqd') # Adjust this to your specific dataset path
```

```python
grouped_values = df_1.groupby(['State_Name', 'District_Name'])['adjusted_margin'].apply(list).reset_index()
# Sort the values by absolute values in descending order within each group
grouped_values['adjusted_margin'] = grouped_values['adjusted_margin'].apply(lambda x: sorted(x, key=lambda v: abs(v), reverse=True))

grouped_values
```

```python
#result_df = df_2.copy()

unique_districts = df_1['District_Name'].unique()
result_df = pd.DataFrame({'District_Name': unique_districts})



max_columns = 20
for i in range(max_columns):
    result_df[f'adjusted_margin_{i + 1}'] = None  # Create empty columns for the values
# Map state information to the result_df based on districts
district_to_state = df_1[['State_Name', 'District_Name']].drop_duplicates().set_index('District_Name')['State_Name']
#district_to_state['PAKAUR']
```

```python
# Fill the result_df with the state and transposed values
for _, row in grouped_values.iterrows():
    state = row['State_Name']
    district = row['District_Name']
    values = row['adjusted_margin']

    # Ensure the district is in the second table
    if district in result_df['District_Name'].values:
        # Insert state information
        result_df.loc[result_df['District_Name'] == district, 'State_Name'] = state
        
        # Insert transposed values into the appropriate columns
        for i, value in enumerate(values):
            result_df.loc[result_df['District_Name'] == district, f'adjusted_margin_{i + 1}'] = value
```

```python

result_df.to_excel('margin_value_2008_new.xlsx', index=False)
```
