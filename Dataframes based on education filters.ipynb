## Join dataframe for each states to create pairs of winner and loser with education filters.

```python
import pandas as pd
table1 = pd.read_excel('2013_win.xlsx',sheet_name='Uttar_Pradesh')
table2 = pd.read_excel('2013_lose.xlsx',sheet_name='Uttar_Pradesh')
```

```python
table1
```

```python
table2 
```

```python
merged = pd.merge(table1, table2, on='Constituency_No')
```

```python
merged
```

```python
# filter merged DataFrame based on S values
result = merged[(merged['educdum_x'] == 0) & (merged['educdum_y'] == 1)]
```

```python
result
```

```python
# select columns from table1
subset = result[[col for col in result.columns if col.endswith('_x')]]
```

```python
subset
```

```python

## import dataframe in same excel and write inside the file saved
import openpyxl
from openpyxl.utils.dataframe import dataframe_to_rows

workbook = openpyxl.load_workbook('final_nongradwin2013.xlsx')

# create a new worksheet
new_sheet = workbook.create_sheet('Uttar_Pradesh')

# write data to the new worksheet
new_data = subset

# Write the DataFrame to the new worksheet
rows = dataframe_to_rows(new_data, index=False, header=True)
for r_idx, row in enumerate(rows, 1):
    for c_idx, value in enumerate(row, 1):
        new_sheet.cell(row=r_idx, column=c_idx, value=value)
        
# save the updated workbook
workbook.save('final_nongradwin2013.xlsx')
```
